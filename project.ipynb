```python
import gradio as gr
from transformers import MarianTokenizer, MarianMTModel

# Load model
model_name = "Helsinki-NLP/opus-mt-en-hi"

tokenizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name)

# Translation function
def translate_text(text):

    tokens = tokenizer(text, return_tensors="pt", padding=True)
    output = model.generate(**tokens)

    translated = tokenizer.decode(output[0], skip_special_tokens=True)

    return translated


# 🔥 Advanced CSS Design
css = """
body{
background: linear-gradient(-45deg,#00c6ff,#0072ff,#00ffcc,#66ff99);
background-size: 400% 400%;
animation: gradientBG 12s ease infinite;
font-family: 'Segoe UI', sans-serif;
}

@keyframes gradientBG{
0%{background-position:0% 50%}
50%{background-position:100% 50%}
100%{background-position:0% 50%}
}

.gradio-container{
background: rgba(255,255,255,0.75);
backdrop-filter: blur(15px);
border-radius:25px;
padding:40px;
box-shadow:0 10px 40px rgba(0,0,0,0.3);
max-width:900px;
margin:auto;
}

h1{
text-align:center;
color:#0f172a;
}

textarea{
font-size:18px !important;
border-radius:14px !important;
border:2px solid #22c55e !important;
padding:10px !important;
}

button{
background:linear-gradient(45deg,#16a34a,#22c55e) !important;
color:white !important;
font-size:18px !important;
border-radius:12px !important;
padding:12px 25px !important;
border:none !important;
}

button:hover{
transform:scale(1.08);
transition:0.3s;
box-shadow:0 6px 20px rgba(0,0,0,0.2);
}

.footer{
text-align:center;
margin-top:20px;
font-size:14px;
color:#333;
}
"""

# Interface Layout
with gr.Blocks(css=css) as app:

    gr.Markdown("""
    # 🌍 AI Language Translator
    ### English ➜ Hindi Translation using Transformers 🤖
    """)

    with gr.Row():

        input_text = gr.Textbox(
            lines=5,
            placeholder="✍️ Type English sentence here...",
            label="English Input"
        )

    with gr.Row():

        output_text = gr.Textbox(
            lines=5,
            label="🇮🇳 Hindi Translation"
        )

    with gr.Row():

        translate_btn = gr.Button("🚀 Translate")
        clear_btn = gr.Button("🗑 Clear")

    translate_btn.click(
        translate_text,
        inputs=input_text,
        outputs=output_text
    )

    clear_btn.click(
        lambda: ("",""),
        None,
        [input_text,output_text]
    )

    gr.Markdown(
        "<div class='footer'>⚡ Powered by Transformers | Built with Gradio</div>"
    )

app.launch()
```
